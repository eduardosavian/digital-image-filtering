```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


from google.colab.patches import cv2_imshow
```

```python
image_name = "t2.jpg"

image_original = cv2.imread('drive/MyDrive/Colab Notebooks/data/' + image_name,1)
```

```python
def clip_image(image: np.ndarray) -> np.ndarray:
    image_clipped = np.clip(image, 0, 255)
    return image_clipped
```

```python
def normalize(image: np.ndarray) -> np.ndarray:
    """Normalize the image and return a numpy array"""
    normalized_image = image / 255.0

    return normalized_image

def denormalize(image: np.ndarray) -> np.ndarray:
    """Denormalize the image and return a numpy array"""
    denormalized_image = (image * 255.0).astype(np.uint8)

    return denormalized_image
```

```python
def convert_to_grayscale_basic(image: np.ndarray) -> np.ndarray:
    grayscale_image = np.dot(image[..., :3], [0.333, 0.333, 0.333]).astype(np.uint8)

    return grayscale_image

grayscale_image = convert_to_grayscale_basic(image_original)
grayscale_image = clip_image(grayscale_image)
cv2_imshow(grayscale_image)
```

```python
def convert_to_grayscale_weighted(image: np.ndarray) -> np.ndarray:
    grayscale_image = np.dot(image[..., :3], [0.299, 0.587, 0.114]).astype(np.uint8)

    return grayscale_image

grayscale_image = convert_to_grayscale_weighted(image_original)
grayscale_image = clip_image(grayscale_image)
cv2_imshow(grayscale_image)
```

```python
def create_histogram_denormalized(image: np.ndarray):
    histogram = np.histogram(image, bins=np.arange(0, 256), range=(0, 256))[0]
    intensity_values = np.arange(0, 256)

    # Make sure both arrays have the same shape
    if len(intensity_values) != len(histogram):
        intensity_values = intensity_values[:-1]  # Remove the last element to match the shape

    plt.bar(intensity_values, histogram, color='blue')
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.show()

create_histogram_denormalized(grayscale_image)
```

```python
create_histogram_denormalized(grayscale_image)
```

```python
def create_histogram_normalized(image: np.ndarray):
    histogram, bin_edges = np.histogram(image, bins=np.arange(0, 256), range=(0, 256), density=True)
    intensity_values = (bin_edges[:-1] + bin_edges[1:]) / 2

    plt.bar(intensity_values, histogram, color='blue')
    plt.xlabel('Intensity')
    plt.ylabel('Normalized Frequency')
    plt.show()
```

```python
create_histogram_normalized(grayscale_image)
```

```python
def convert_to_negative(image: np.ndarray) -> np.ndarray:
    image_negative = 255 - image

    return image_negative

image_negative = convert_to_negative(image_original)
image_negative = clip_image(image_negative)
cv2_imshow(image_negative)
```
